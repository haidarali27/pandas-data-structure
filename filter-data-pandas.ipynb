```python
# read data source first
data_supermarket = pd.read_csv('supermarket_sales.csv')
data_supermarket.head(2)
```

```python
#filter data based on value
filter_data = data_supermarket[data_supermarket["Payment"] == "Ewallet"]
# print data
filter_data.head(2)
```

```python
# filter data berdasarkan beberapa kondisi yang harus dipenuhi
filter_data2 = data_supermarket[(data_supermarket["Product line"]== "Health and beauty") | (data_supermarket["Product line"]== "Electronic accessories")]
filter_data2.head(10)
```

```python
# filter data based on data yang ditentukan
filter_data3 = data_supermarket[data_supermarket["Unit price"]> 50.50]
filter_data3.head(10)
```

```python
#filter dengan nilai yang sudah ditentukan lebih dari 2
filter_data4 = data_supermarket[(data_supermarket["Total"] <= 300) & (data_supermarket["Rating"] >= 7.5)]
filter_data4.head(10)
```
